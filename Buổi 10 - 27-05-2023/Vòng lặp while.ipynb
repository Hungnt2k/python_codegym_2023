## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
# n = 10
n = int(input("Input n: "))
i = 0  # thiết lập biến đếm ban đầu

while i <= n:  # điều kiện vào lặp
    print(i)
    i += 2  # cập nhật biến đếm sau mỗi lần lặp
    
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
n = int(input("input n: "))
i = 1
while n >= i:
    if n % 2 != 0:
        print(n)
    n -= 1
```

### Ví dụ: Đảo ngược chuỗi

```python
n = str(input("Input string: "))
i = len(n) -1 

while i >= 0:
    print(n[i], end="")
    i -= 1
    
```

### Ví dụ: Tam giác vuông

```python
"""
Please input the height of triangle: 5
*
**
***
****
*****

"""

n = int(input("Please input the height of triangle: "))
i = "*"

while len(i) <= n:
    print(i)
    i += "*"
```

### Ví dụ: Vẽ đa giác đều

```python
import turtle
size = int(input("Nhập độ dài một cạnh đa giác: "))
sides = int(input("Nhập số cạnh đa giác: "))
n = turtle.Turtle()
i = 0
while i < sides:
    n.forward(size)
    n.right(360/sides)
    i = i + 1
turtle.done()
```

### Ví dụ: 
Nhập vào một số nguyên a

Cho S(k) = 1 + 2 + 3 + ... + k

Lựa chọn k sao cho S(k) > a với k nhỏ nhất

In S(k) và k lên màn hình

```python
a = int(input("Nhập vào số nguyên a: "))

k = 1
sum_k = 0


while sum_k <= a:
    sum_k += k
    k += 1
    
print("S(k) =", sum_k)
print("k =", k-1)
```
